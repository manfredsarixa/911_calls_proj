```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('whitegrid')
```

```python
df = pd.read_csv('911.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df['zip'].value_counts().head(5) #top 5 zipcodes for 911 calls
```

```python
df['twp'].value_counts().head(5) # top 5 townships (twp) for 911 calls
```

```python
df['title'].nunique() #no. of unique title codes
```

```python
df['Reason'] = df['title'].apply(lambda x: x.split(':')[0]) 
# creating a new column called "Reason"
# using lambda function to split the title column and select the first element
# E.g. if the title column value is EMS: BACK PAINS/INJURY , the Reason column value would be EMS
```

```python
df['Reason'].value_counts() #most common Reason for a 911 call based off of the new col created
```

```python
sns.countplot(x='Reason',data=df,palette='viridis') # countplot of 911 calls by Reason
```

```python
type(df['timeStamp'].iloc[0]) # checking the data type of timeStamp column 
```

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp']) # converting the timeStamp column from strings to datetime object
```

```python
df['Hour'] = df['timeStamp'].apply(lambda x: x.hour)
df['Month'] = df['timeStamp'].apply(lambda x: x.month)
df['Day Of Week'] = df['timeStamp'].apply(lambda x: x.dayofweek)
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun',} # mapping the actual string names to the day of the week
```

```python
df['Day Of Week'] = df['Day Of Week'].map(dmap)
```

```python
sns.countplot(x='Day Of Week',data=df,hue='Reason',palette='viridis')

# To relocate the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sns.countplot(x='Month',data=df,hue='Reason',palette='viridis')

# To relocate the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
byMonth = df.groupby('Month').count()
byMonth.head()
# creating a gropuby object called byMonth
#group the DataFrame by the month column and use the count() method for aggregation
```

```python
byMonth['twp'].plot() # plotting the count of 911 calls per month
```

** Now see if you can use seaborn's lmplot() to create a linear fit on the number of calls per month. Keep in mind you may need to reset the index to a column. **

```python
sns.lmplot(x='Month',y='twp',data=byMonth.reset_index())
```

**Create a new column called 'Date' that contains the date from the timeStamp column. You'll need to use apply along with the .date() method. ** 

```python
df['Date'] = df['timeStamp'].apply(lambda t: t.date())
```

** Now groupby this Date column with the count() aggregate and create a plot of counts of 911 calls.**

```python
df.groupby('Date').count()['twp'].plot()
plt.tight_layout()
```

** Now recreate this plot but create 3 separate plots with each plot representing a Reason for the 911 call**

```python
df[df['Reason']=='Traffic'].groupby('Date').count()['twp'].plot()
plt.title('Traffic')
plt.tight_layout()
```

```python
df[df['Reason']=='Fire'].groupby('Date').count()['twp'].plot()
plt.title('Fire')
plt.tight_layout()
```

```python
df[df['Reason']=='EMS'].groupby('Date').count()['twp'].plot()
plt.title('EMS')
plt.tight_layout()
```

____
** Now let's move on to creating  heatmaps with seaborn and our data. We'll first need to restructure the dataframe so that the columns become the Hours and the Index becomes the Day of the Week. There are lots of ways to do this, but I would recommend trying to combine groupby with an [unstack](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.unstack.html) method. Reference the solutions if you get stuck on this!**

```python
DayHour = df.groupby(by=['Day Of Week','Hour']).count()['Reason'].unstack()
DayHour.head()
```

** Now create a HeatMap using this new DataFrame. **

```python
plt.figure(figsize=(12,6))
sns.heatmap(DayHour,cmap='viridis')
```

** Now create a clustermap using this DataFrame. **

```python
sns.clustermap(DayHour,cmap='coolwarm')
```

** Now repeat these same plots and operations, for a DataFrame that shows the Month as the column. **

```python
DayMonth = df.groupby(by=['Day Of Week','Month']).count()['Reason'].unstack()
DayMonth.head()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(DayMonth,cmap='viridis')
```

```python
sns.clustermap(DayMonth,cmap='viridis')
```
